The aim is to show that the cost of smc is larger than the cost of sgd at the same accuracy level

```python
def costs(m,s,epsilon,
          sig_th,sig_x,
          func, f, g,
          initial,
          n_experiments=int(10**4),
          n=10):
    
    T=int(3*np.log(1/epsilon)/m)+1 
    # the terminal time
    
    h_list=[4*epsilon/2**i/m for i in range(1,10)] 
    # the list of stepsize candidates in an descending order
    print("The h candidate list we use is",h_list)
    
    euler_error_list=[]
    sgd_error_list=[]
    euler_cost_list=[]
    sgd_cost_list=[]
    
    for j in range(n):
        
    #  n many experiments       
            print("the ", j, "-th experiment:")
            theta = sig_th*np.random.randn(1);
            X = theta + np.random.randn(m)*sig_x
            #     data generation
            var_post, mu_post=var_mu_comp(X,sig_th,sig_x)
            
            print("the posterior variance and mean",var_post, mu_post)
            ans=ref_im_sol(mu_post,var_post,func)
            print("the ground truth is",ans)
            
            #  now try to find a suitable stepsize
            i=0
            x1,v1=functional_euler_generator(n_experiments,T,
                                                       h_list[i], 
                                                       f,
                                                       g,
                                                       initial,
                                                       mu_post, 
                                                       var_post,  
                                                       func)


            x2,v2=functional_euler_generator(n_experiments,
                                                   T,
                                                   h_list[i+1], 
                                                   f,
                                                   g,
                                                   initial,
                                                   mu_post, 
                                                   var_post,  
                                                   func)  
            
#             the creteria is to check prediction of consective stepsize
            while (np.abs(x1-x2)>epsilon) | np.isnan(v1) | np.isnan(v2):


                    x1,v1=x2,v2
                    i=i+1
                    x2,v2=functional_euler_generator(n_experiments,T,
                                                   h_list[i+1], 
                                                   f,
                                                   g,
                                                   initial,
                                                   mu_post, 
                                                   var_post,  
                                                   func)


            h=h_list[i]/2 
#             the final choice of stepsize
            print("The stepsize we will use is the ", i+1, "th value in the candidate list=", h)
    
            N_full=int(2*epsilon**(-2)*v2)+1
# the number of sample for Euler
            n_step=int(T/h)
            

            start_time=time.time()
            full_euler_sol,_=functional_euler_generator(N_full,T,
                                                       h, 
                                                       f,
                                                       g,
                                                       initial,
                                                       mu_post, 
                                                       var_post,  
                                                       func)
            end_time=time.time()

            euler_error_list.append((full_euler_sol-ans)**2)
            euler_cost_list.append(end_time-start_time)
            
            N_sgd=int(1/epsilon**2/m)+1
# the number of sample for SGD, where the stepsize shared with Euler

            print("the number of samples for two methods", N_full,N_sgd)
            start_time=time.time()
            full_sgd_sol,_=functional_sgd_generator(N_sgd,T,
                                                       h, 
                                                       f,
                                                       g,
                                                       initial,
                                                       s,
                                                       sig_th,
                                                       sig_x,
                                                       X,                                                      
                                                       mu_post, 
                                                       var_post,  
                                                       func)
            end_time=time.time()  

            sgd_error_list.append((full_sgd_sol-ans)**2)
            sgd_cost_list.append(end_time-start_time)
            
            print("\n")
        
    return np.sqrt(np.mean(euler_error_list)),\
           np.mean(euler_cost_list),\
           np.sqrt(np.mean(sgd_error_list)),\
           np.mean(sgd_cost_list)
```

```python
m,s,epsilon=10**3,2,1/10

sig_th, sig_x=0.2, 0.15

initial=1
g=1

func = lambda y :np.sin(y)+np.abs(y)+1 # functional

costs(m,s,epsilon,
      sig_th, sig_x,
      func,f, g,
      initial,
      n_experiments=int(10**4),
      n=10)
```
